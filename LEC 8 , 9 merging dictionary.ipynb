# Score processor fungtion

```python
def get_scores_df(report_card_dictionary,division_name):
    
    # convert dictionaryinto data frame
    report_card_df = pd.DataFrame(report_card_dictionary)
    
    # add total score column
    report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["bio_scores"] + report_card_df["history_score"] + report_card_df["physics_scores"]
    
    # add average score column
    report_card_df["average_score"] = report_card_df["total_score"]/4
    
    # add pass fail column
    report_card_df.loc[report_card_df["total_score"]>=140, "status"] = "pass"
    report_card_df.loc[report_card_df["total_score"]<140, "status"] = "Fail"
    
    # add character grade column
    report_card_df.loc[report_card_df["average_score"]>=90, "character_grade"] = "A"
    report_card_df.loc[(report_card_df["average_score"]>=70)&(report_card_df["average_score"]<90), "character_grade"] = "B"
    report_card_df.loc[(report_card_df["average_score"]>=50)&(report_card_df["average_score"]<70), "character_grade"] = "C"
    report_card_df.loc[(report_card_df["average_score"]>=35)&(report_card_df["average_score"]<50), "character_grade"] = "D"
    report_card_df.loc[report_card_df["average_score"]<35, "character_grade"] = "F"
    
    
    # add division column
    report_card_df["division"] = division_name
    
    # return report card data frame
    return report_card_df
```

```python
div_a_score_card = get_scores_df(div_a_score_dictionary, "A")
div_a_score_card
```

# Get email df fungtion

```python
def get_email_df(emails_dictionary,division_name):
    
    # converting our dictionary into a data frame
    emails_df = pd.DataFrame(emails_dictionary)
    
    # add a division column to our data frame
    emails_df["division"] = division_name
    
    #converting all emails to lower case
    emails_df["student_email"] = emails_df["student_email"].str.lower()
    
    #return email df
    return emails_df
```

```python
div_a_email_df = get_email_df(div_a_email_dictionary, "A")
div_a_email_df
```

# Get scores and email df fungtion

```python
def get_scores_email_df(scores_dictionary,emails_dictionary,division_name):
    
    # Step 1 : make a score dictionary into a scorecard df
    score_card_df = get_scores_df(scores_dictionary, division_name)
    
    # Step 2 : make the email dictionary into a email dataframe
    emails_df = get_email_df(emails_dictionary, division_name)
    
    # Step 3 : merge the score with email df
    score_card_email_df = pd.merge(score_card_df,emails_df,how = "left", on = ["student_names","division"])
    
    # Step 4 : return merged df
    return score_card_email_df
```

```python
get_scores_email_df(div_a_score_dictionary, div_a_email_dictionary, "A")
```

# Get all divisions DF fungtion

```python
def get_all_divisions_df(div_1_scores_dictionary,div_1_emails_dictionary, div_1_name,
                        div_2_scores_dictionary,div_2_emails_dictionary, div_2_name,
                        div_3_scores_dictionary,div_3_emails_dictionary, div_3_name):
    
    div_1_df = get_scores_email_df(div_1_scores_dictionary,div_1_emails_dictionary, div_1_name)
    div_2_df = get_scores_email_df(div_2_scores_dictionary,div_2_emails_dictionary, div_2_name)
    div_3_df = get_scores_email_df(div_3_scores_dictionary,div_3_emails_dictionary, div_3_name)
    
    
    all_division_df = pd.concat([div_1_df,div_2_df,div_3_df], ignore_index=True)
    
    return all_division_df
```

## Save all division DF fungtion

```python
def get_save_divisions_df(div_1_scores_dictionary,div_1_emails_dictionary, div_1_name,
                        div_2_scores_dictionary,div_2_emails_dictionary, div_2_name,
                        div_3_scores_dictionary,div_3_emails_dictionary, div_3_name):
    
    div_1_df = get_scores_email_df(div_1_scores_dictionary,div_1_emails_dictionary, div_1_name)
    div_2_df = get_scores_email_df(div_2_scores_dictionary,div_2_emails_dictionary, div_2_name)
    div_3_df = get_scores_email_df(div_3_scores_dictionary,div_3_emails_dictionary, div_3_name)
    
    
    all_division_df = pd.concat([div_1_df,div_2_df,div_3_df], ignore_index=True)
    
    all_division_df.to_csv("all Division scorecard.csv")
```
